```python
import numpy as np
import pandas as pd

from sklearn.neighbors import KNeighborsClassifier
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
```

```python
train_data = pd.read_csv('train.csv', sep = ",")

train_data = train_data.drop('PassengerId', axis = 1)
train_data.loc[train_data["Sex"] == "male", "Sex"] = 1
train_data.loc[train_data["Sex"] == "female", "Sex"] = 0
train_data.head()
```

```python
test_data = pd.read_csv('test.csv', sep = ",")

test_data = test_data.drop('PassengerId', axis = 1)
test_data.loc[test_data["Sex"] == "male", "Sex"] = 1
test_data.loc[test_data["Sex"] == "female", "Sex"] = 0
test_data.head()
```

```python
y_train = train_data['Survived']
X_train = train_data.drop(["Name", "Cabin", "Embarked", "Survived", "Ticket"], axis = 1)
X_test = test_data.drop(["Name", "Cabin", "Embarked", "Ticket"], axis = 1);
```

```python
logreg = LogisticRegression()
logreg.fit(X_train, y_train)
y_pred = logreg.predict(X_test)
logreg.score(X_train, y_train)
```

```python
knn = KNeighborsClassifier(n_neighbors = 5)
knn.fit(X_train, y_train)
y_pred = knn.predict(X_test)
knn.score(X_train, y_train)
```

```python
random_forest = RandomForestClassifier(n_estimators=100)
random_forest.fit(X_train, y_train)
y_pred = random_forest.predict(X_test)
random_forest.score(X_train, y_train)
```
